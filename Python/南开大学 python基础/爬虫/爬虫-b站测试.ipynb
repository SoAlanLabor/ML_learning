```python
from selenium import webdriver
import time
from selenium.webdriver.edge.options import Options
from selenium.webdriver.common.by import By

class CtripCommentsCrawler:
    def __init__(self, url, keyword=None):
        self.url = url
        self.keyword = keyword
        self.sum = 0.0
        self.author_sum = 0.0
        self.author_segment = 0.0
        self.titles = None

    def setZero(self):
        self.sum = 0.0
        self.author_sum = 0.0
        self.author_segment = 0.0

    def getdata(self, pagenum):
        try:
            option = Options()
            option.add_experimental_option('excludeSwitched',['enable-automation'])

            driver = webdriver.Edge()
            driver.maximize_window()
            driver.get(self.url)

            for pageno in range(1,pagenum+1):
                self.setZero()
                trynum = 0
                while True:
                    try:
                            self.titles = driver.find_elements(By.CSS_SELECTOR,"#i_cecream > div > div:nth-child(2) > div.search-content > div > div > div.video-list.row > div")
                            break
                    except:
                            trynum += 1
                            if trynum > 6:
                                break
                            time.sleep(5)
                            continue
                for i in range(0,29):
                    title_main = self.titles[i].find_element(By.CLASS_NAME,"bili-video-card__info--tit").text

                    congratulation_0 = self.titles[i].find_element(By.CLASS_NAME, "bili-video-card__stats--item")
                    congratulation_1 = congratulation_0.find_element(By.TAG_NAME,"span").text

                    author = self.titles[i].find_element(By.CLASS_NAME,"bili-video-card__info--author").text

                    date_0 = self.titles[i].find_element(By.CLASS_NAME, "bili-video-card__info--date").text
                    print('{0} 作者：{1} 该视频播放量为{2},发布日期是{3}'.format(title_main, author,congratulation_1, date_0[2:]))
                    self.sum += eval(congratulation_1[:-1])

                    if author == self.keyword:
                        self.author_sum += eval(congratulation_1[:-1])

                self.author_segment = 100 * self.author_sum / self.sum
                print("全程总计播放量:%.1f万" %self.sum)
                print(self.keyword +"视频的全程总计播放量:%.1f万,总量占比为:%0.1f%%" % (self.author_sum,self.author_segment))
                print('-----page {0}-----'.format(pageno))

                try:
                    nextpage = driver.find_element(By.CSS_SELECTOR,'#i_cecream > div > div:nth-child(2) > div.search-content > div > div > div.flex_center.mt_x50.mb_lg > div > div > button:nth-child(11)')
                    driver.execute_script("arguments[0].click();", nextpage)
                    time.sleep(5)
                except:
                    break
        finally:
            driver.quit()

if __name__=='__main__':
    url = 'https://search.bilibili.com/all?vt=72366432&keyword=%E5%8D%97%E5%BC%80%E5%A4%A7%E5%AD%A6&from_source=webtop_search&spm_id_from=333.1007&search_source=5&order=click'
    crawler = CtripCommentsCrawler(url,"南开大学")
    crawler.getdata(10)
```
